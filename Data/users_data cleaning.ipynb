```python
import pandas as pd
```

```python
user1 = pd.read_csv('final_user_0616_0630.csv',lineterminator='\n')
user2 = pd.read_csv('users_2022-07-01.csv')
user3 = pd.read_csv('keywords2_users_0617-0701.csv')
```

```python
print(len(user1))
print(len(user2))
print(len(user3))
```

```python
users = pd.concat([user1,user2,user3],axis=0)
users.shape
```

```python
users.head()
```

```python
users = users.drop_duplicates(subset=['id'])
users.shape
```

```python
users['created'] = pd.to_datetime(users['created'], errors='coerce')
```

```python
users = users.dropna(subset=['username','id'])
```

```python
users.shape
```

```python
users.to_csv('abortion_users0617-0701.csv', index=False)
```
