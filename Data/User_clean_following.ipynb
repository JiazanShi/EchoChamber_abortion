For this project, we only need the following relationship among users who joined the online discussion about abortion. This notebook is used for cleaning following list

```python
import pandas as pd

data = pd.read_csv('user_following.csv')
data.head()
```

```python
data['following'] = data['following'].astype('str')
```

```python
# Pandas saves lists inside of dataframes as Strings.
#This function parses it back into a list of username.
def parse_following_string(s):
    s = s[1:len(s) - 2]
    ids = []
    for id in s.split(', '):
        if id != '':
            ids.append(id.strip("''"))
    return ids
```

```python
parse_following_string(data['following'][0])
```

```python
'dom_lucre' in parse_following_string(data['following'][0])
```

```python
#use set to save all the unique username
user_set = set(data.username)
```

```python
data['clean_following'] = None
data.head()
```

```python
#clean following list and only keep users who are in the user set
for i in range(0,len(data)):
    username = data['username'][i]
    clean_following = []
    following = parse_following_string(data['following'][i])
    for follower in following:
        if follower in user_set:
            clean_following.append(str(follower))
    data['clean_following'][i] = clean_following
```

```python
data.head()
```

```python
data['following_count'] = None
for i in range(len(data)):
    data['following_count'][i] = len(data['clean_following'][i])
```

```python
#filter out users who don't following anyone else
data.loc[data['following_count']!=0]
```

```python
data.to_csv('following_clean.csv',index=False)
```
