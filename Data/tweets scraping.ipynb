```python
import pandas as pd
import snscrape.modules.twitter as twitter
import dill
```

```python
keywords = '"abortion" OR "roe wade" OR "pro choice" OR "pro life" OR "interrupt pregnancy" OR "planned parenthood"'
```

```python
#from 2022-06-17 to 2022-07-01
#1: 0617-0621
#2: 0621-0624
#3: 0624-0626
#4: 0626 - 0702
date_start = '2022-06-26'
date_end = '2022-07-02' #end date is not inclusive
```

```python
tweets_list = []
user_dict = dict() #created the empty dictionary for all user info
```

```python
for tweet in twitter.TwitterSearchScraper(keywords  + 'lang:en since:' + date_start + ' until:' + date_end).get_items():
    if tweet.user.id not in user_dict:
        user_dict[tweet.user.id] = tweet.user

    #return all columns    
    tweets_list.append([tweet.url,
                        tweet.date,
                        tweet.rawContent,
                        tweet.renderedContent,
                        tweet.id,
                        #store the user id, the dict in tweet.user cannot be converted to csv
                        tweet.user.id,
                        tweet.replyCount,
                        tweet.retweetCount,
                        tweet.likeCount,
                        tweet.quoteCount,
                        tweet.conversationId,
                        tweet.lang,
                        tweet.source,
                        tweet.sourceUrl,
                        tweet.sourceLabel,
                        tweet.links,
                        tweet.media,
                        tweet.retweetedTweet,
                        tweet.quotedTweet,
                        tweet.inReplyToTweetId,
                        tweet.inReplyToUser,
                        tweet.mentionedUsers,
                        tweet.coordinates,
                        tweet.place,
                        tweet.hashtags,
                        tweet.cashtags,
                        tweet.card,
                        tweet.viewCount,
                        tweet.vibe,
                        tweet.content,
                        tweet.outlinks,
                        tweet.outlinksss,
                        tweet.tcooutlinks,
                        tweet.tcooutlinksss,
                        tweet.username])
    if len(tweets_list) % 10000 == 0:
        print(len(tweets_list)) #get the progress

# Save the data immediately, or you may risk loosing the data if the kernel crashes
tweets=pd.DataFrame(tweets_list,columns = ['url', 'date', 'rawContent', 'renderedContent', 'id', 'userid',
       'replyCount', 'retweetCount', 'likeCount', 'quoteCount',
       'conversationId', 'lang', 'source', 'sourceUrl', 'sourceLabel', 'links',
       'media', 'retweetedTweet', 'quotedTweet', 'inReplyToTweetId',
       'inReplyToUser', 'mentionedUsers', 'coordinates', 'place', 'hashtags',
       'cashtags', 'card', 'viewCount', 'vibe', 'content', 'outlinks',
       'outlinksss', 'tcooutlinks', 'tcooutlinksss', 'username'])

tweets.to_csv('tweets_' + date_start  + '_' + date_end + '.csv', index = False, encoding='utf-8-sig')

dill.dump(user_dict, open('2_user_dict_' + date_start + '_' + date_end + '.dat', 'wb'))
#users=pd.DataFrame(user_list,columns = ['id', 'username', 'rawDescription', 'verified'])
#users.to_csv('users_' + date_start + '_' + date_end + '.csv', index = False, encoding='utf-8-sig')
```

```python
#convert dict to dataframe and save user info as csv
userdf = pd.DataFrame.from_dict(user_dict, orient='index')
userdf.to_csv('users_' + date_start + '_' + date_end + '.csv', index = False, encoding='utf-8-sig')
```
