```python
import pandas as pd
```

```python
data1 = pd.read_csv('tweets_0616-0630.csv',lineterminator='\n')
data2 = pd.read_csv('tweets_2022-07-01.csv')
data3 = pd.read_csv('keywords2_tweets_0617-0701.csv')
```

```python
print(len(data1))
print(len(data2))
print(len(data3))
```

```python
tweets = pd.concat([data1,data2,data3],axis=0)
tweets.shape
```

```python
len(tweets['id'].unique())
```

```python
new_tweets = tweets.drop_duplicates(subset=['id'])
```

```python
new_tweets.shape 
```

```python
new_tweets.head()
```

```python
new_tweets['date'] = pd.to_datetime(new_tweets['date'], errors='coerce')
```

```python
new_tweets['date_day'] = new_tweets['date'].dt.date
new_tweets.head()
```

```python
new_tweets['date_day'].value_counts()
```

```python
new_tweets = new_tweets.loc[new_tweets['date_day']!=pd.to_datetime('2022-06-16')]
```

```python
new_tweets.shape
```

```python
new_tweets = new_tweets.dropna(subset=['username', 'date','rawContent','userid','id','content'])
new_tweets.shape
```

```python
new_tweets.sort_values(by = 'date', inplace= True)
```

```python
new_tweets.reset_index(inplace= True)
```

```python
new_tweets = new_tweets.drop(columns = ['index','date_day'])
```

```python
new_tweets.shape
```

```python
new_tweets.head()
```

```python
new_tweets.to_csv('abortion_tweets0617-0701.csv',index=False)
```
